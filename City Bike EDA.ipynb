# Q. Which bike stations are most popular to start trips from?

```python
df2 = df.groupby(["start station id"]).size().reset_index(name='counts')
df2
```

```python
temp = df.drop_duplicates('start station id')
# left join
df2 = pd.merge(df2, temp[['start station id', 'start station name','start station latitude',
       'start station longitude']], how = 'left', on= ['start station id'])
```

```python
import folium
```

```python
#initialize empty map
m = folium.Map(location=[40.691966,-73.981302], titles = 'OpenStreetMap',zoom_start=14)
```

```python
m
```

```python
for i in range(0,len(df2)):
    folium.Circle(
        location = [df2.iloc[i]['start station latitude'],df2.iloc[i]['start station longitude']],
        radius = float(df2.iloc[i]['counts'])/2,
        color = 'green',
        fill = True, fill_color = 'green'
    ).add_to(m)
```
